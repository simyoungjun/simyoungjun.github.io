### Analyze Data

```python
from dataclasses import dataclass
from typing import Literal
import os

# fields with None will be not used when tech is gt
@dataclass
class WaveItem:
    division1: Literal["m2m", "a2a"]
    division2: Literal["m2m", "f2f", "m2f", "f2m", "en_m2kr_f", "en_f2fr_m"]
    tech: Literal["gt", "acvc", "autovc", "againvc", "vqvc"]
    src_lang: Literal["en", "kr", "fr"]
    src_spk: str
    src_wav: str | None
    tgt_spk: str | None
    tgt_wav: str | None
    my_wav: str
    id: str

items = []
for colsi, cols in enumerate(notion_html_meta_cols):
    division1 = notion_html_meta_cols_dim1[colsi][0]
    src_lang = notion_html_meta_cols_dim1[colsi][1]
    division2_list = notion_html_meta_cols_dim1[colsi][2]
    src_spk_list = notion_html_meta_cols_dim1[colsi][3]
    tgt_spk_list = notion_html_meta_cols_dim1[colsi][4]
    for coli, col in enumerate(cols):
        tech_full = notion_html_meta_cols_dim2[coli]
        if tech_full == "gt_tgt":
            continue
        tech = tech_full.split("_")[0]
        for div2i, division2 in enumerate(division2_list):
            src_spk = src_spk_list[div2i] if tech_full != "gt_tgt" else tgt_spk_list[div2i]
            tgt_spk = tgt_spk_list[div2i] if tech != "gt" else None
            src_wav = None
            tgt_wav = None
            if tech != "gt":
                src_wav = notion_html_data[col][0]
                tgt_wav = notion_html_data[col][1]
            items.append(
                WaveItem(
                    division1=division1,
                    division2=division2,
                    tech=tech,
                    src_lang=src_lang,
                    src_spk=src_spk,
                    src_wav=src_wav,
                    tgt_spk=tgt_spk,
                    tgt_wav=tgt_wav,
                    my_wav=notion_html_data[col][div2i],
                    id=os.urandom(16).hex()
                )
            )
```

### Filter Wave Items for Similarity MOS

```python
items_by_tgt_spk: dict[str, list[WaveItem]] = {}
for item in items:
    tgt_spk = item.tgt_spk
    if tgt_spk is None:
        if item.tech != "gt":
            raise ValueError("tgt_spk is None but tech is not gt")
        tgt_spk = item.src_spk
    if tgt_spk not in items_by_tgt_spk:
        items_by_tgt_spk[tgt_spk] = []
    items_by_tgt_spk[tgt_spk].append(item)

items_by_tgt_spk
```

```python
import random

def sim_filter(spk: str, spk_items: list[WaveItem]) -> list[WaveItem]:
    do_it = False
    do_it2 = False

    for item in spk_items:
        if item.tech == "gt":
            do_it = True
            continue
        if item.tech != "gt":
            do_it2 = True
            continue

    if (not do_it) or (not do_it2):
        return []
    
    # shuffle items
    spk_items = spk_items.copy()
    random.shuffle(spk_items)

    new_list = []
    added_techs = []
    gt_lang = ""
    # add gt
    for item in spk_items:
        if item.tech == "gt":
            new_list.append(item)
            added_techs.append(item.tech)
            gt_lang = item.src_lang
            break
    assert gt_lang != ""
    # add items with same language
    for item in spk_items:
        if item.tech not in added_techs:
            if item.src_lang == gt_lang:
                new_list.append(item)
                added_techs.append(item.tech)
    # add items with different language
    for item in spk_items:
        if item.tech not in added_techs:
            new_list.append(item)
            added_techs.append(item.tech)
    return new_list


sim_items = {k: sim_filter(k, v) for k, v in items_by_tgt_spk.items()}

# remove empty items
sim_items = {k: v for k, v in sim_items.items() if len(v) > 0}

sim_items
```

### Generate Data for Web Page

```python
import os
import shutil

shutil.rmtree("./data", ignore_errors=True)
os.makedirs("./data", exist_ok=True)

def copy_wav():
    copied_ids = []
    for item in nat_items:
        if item.id in copied_ids:
            continue
        shutil.copy("./src/" + item.my_wav, "./data/" + item.id + ".wav")
        copied_ids.append(item.id)
    for sim_group_items in sim_items.values():
        for item in sim_group_items:
            if item.id in copied_ids:
                continue
            shutil.copy("./src/" + item.my_wav, "./data/" + item.id + ".wav")
            copied_ids.append(item.id)

copy_wav()
```

```python
import json

web_dict = {}

for nati, nat_item in enumerate(nat_items):
    spk_info = f"{nat_item.src_spk} to {nat_item.tgt_spk}" if nat_item.tgt_spk is not None else nat_item.src_spk
    web_dict[f"n{nati}"] = [f"./data/{nat_item.id}.wav", nat_item.src_lang, f"{nat_item.tech} | {nat_item.division1} | {nat_item.division2} | {spk_info}"]

for simi, (sim_spk, sim_group_items) in enumerate(sim_items.items()):
    simi_gt = ""
    simi_lang = ""
    simi_comp = []
    for sim_item in sim_group_items:
        if sim_item.tech == "gt":
            if simi_gt != "":
                print(f"multiple gt in one group: {sim_spk}")
            simi_gt = sim_item.id
            simi_lang = sim_item.src_lang
            continue
        spk_info = f"{sim_item.src_spk} ({sim_item.src_lang}) to {sim_item.tgt_spk}"
        simi_comp.append([f"./data/{sim_item.id}.wav", f"{sim_item.tech} | {sim_item.division1} | {sim_item.division2}", spk_info])
    if simi_gt == "" or len(simi_comp) == 0:
        print(f"simi_gt or simi_comp is empty: {sim_spk}")
        continue
    else:
        web_dict[f"s{simi}"] = [f"./data/{simi_gt}.wav", simi_comp, simi_lang, sim_spk]


with open("./data.json", "w") as f:
    json.dump(web_dict, f)
```
